```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.shape
titanic.head()
```

```python
for c in titanic.columns:
    print(c, '\n', titanic[c].describe(), '\n', '\n')
    try:
        titanic[c].plot.kde()
        plt.show()
    except:
        # will only work with numeric data
        pass
```
